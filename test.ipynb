```python
import numpy as np
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data1= 100 * np.random.random(100)
data2 = data1 + 10 + np.random.random(100)
data3 = 100 * np.random.random(100)
data4 = data2 + data3 + np.random.random(100) - 8

d = np.vstack([data1, data2, data3, data4]).T
```

```python
df = pd.DataFrame(d, columns=['data1', 'data2', 'data3', 'data4'])
df
```

```python
Z = scipy.cluster.hierarchy.linkage(df.corr(), 'single', 'correlation')

fig, ax = plt.subplots(figsize=(10, 10))
scipy.cluster.hierarchy.dendrogram(Z, color_threshold=0);
```

```python
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram
from scipy.spatial.distance import pdist, squareform

D = squareform(pdist(d.T, 'correlation'))
h = sns.clustermap(D, cmap='Reds')
```

```python
Z = h.dendrogram_col.linkage    
dendrogram(Z, color_threshold=0)
```
